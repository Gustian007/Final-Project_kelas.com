```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import statistics as stats

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
house = pd.read_csv(r"D:\train_house - train_house.csv")
house.head()
```

# Check & Drop NaN

```python
total = house.isnull().sum().sort_values(ascending=False)
percent = (house.isnull().sum()/house.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
house = house.dropna(axis=1)
house.head()
```

# Memisahkan Data numerik dan kategorikal

## Numerik

```python
num_cols = house._get_numeric_data().columns
house_numeric = house[num_cols]
house_numeric.head()
```

## Katerogikal

```python
house_categorical = house.drop(num_cols, axis = 1)
house_categorical.head()
```

```python
house_numeric.to_csv('house_numeric.csv')
house_categorical.to_csv('house_categorical.csv')
```

```python
house_categorical = pd.read_csv('house_categorical.csv',index_col=0)
house_numerical = pd.read_csv('house_numeric.csv',index_col=0)
```

# 1D Visualization (Kategorikal)

```python
sns.barplot(x=house_categorical['Utilities'].value_counts().index,
                 y=house_categorical['Utilities'].value_counts(normalize = True))
plt.title('Utilities')
plt.show()
```

### Analisa

```python
sns.barplot(x=house_categorical['HouseStyle'].value_counts().index,
                 y=house_categorical['HouseStyle'].value_counts(normalize = True))
plt.title('HouseStyle')
plt.show()
```

### Analisa
Dari barplot tersebut dapat dilihat bahwa:

## Exterior Quality

```python
sns.barplot(x=house_categorical['ExterQual'].value_counts().index,
                 y=house_categorical['ExterQual'].value_counts(normalize = True))
plt.title('ExterQual')
plt.show()
```

### Analisa

# 1D Visulization (Numerik)

## SalePrice

```python
x = house_numerical["SalePrice"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

### Analisa  
- Rata-rata (Mean): \$180,921  
- Median: \$163,000  
- Standar deviasi: \$79,443  

Distribusi menunjukkan Right Skewed, artinya:
- Mayoritas harga rumah berada di bawah rata-rata.
- Beberapa rumah mahal mendorong nilai rata-rata lebih tinggi dari median.


## GrLivArea

```python
x = house_numerical["GrLivArea"]
sns.distplot(x, axlabel=x.name, bins = 100)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

### Analisa

- Distribusi GrLivArea juga menunjukkan positive skewness (kanan).
- Mean (1515) lebih besar dari Median (1464).
- Terdapat beberapa rumah dengan luas sangat besar (> 4000 sqft).
- Sebagian besar rumah memiliki living area antara 1000â€“2000 sqft.
- Penyebaran cukup tinggi (standar deviasi â‰ˆ 525), menunjukkan variasi signifikan dalam ukuran rumah.


```python
x = house_numerical["GarageArea"]
sns.distplot(x, axlabel=x.name, bins = 100)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

### Analisa

- Distribusi `GarageArea` menunjukkan **sedikit skew ke kanan (positive skewness)**.
- **Median = 480**, sedikit lebih besar daripada **Mean = 472.98**.
- Sebagian besar rumah memiliki **luas garasi antara 200â€“700 sqft**.
- Ada beberapa outlier dengan **luas garasi > 1000 sqft** yang menyebabkan ekor ke kanan.
- Penyebaran (standar deviasi â‰ˆ 213) cukup besar, mengindikasikan variasi ukuran garasi antar rumah.
- Terdapat puncak tajam di area 0, kemungkinan menunjukkan rumah tanpa garasi.


# Pemrosesan Data Variabel Numerikal

## Sale Price

```python
x = np.log10(house_numerical["SalePrice"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

## GrLivArea

```python
x = np.log10(house_numerical["GrLivArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

## Garage Area

```python
# Menghilangkan nilai 0

x = house_numerical["GarageArea"]
x = x[x != 0]

sns.distplot(x, axlabel=x.name, bins=100)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label="mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label="median")
plt.legend(handles=[line1, line2], loc=1)
plt.title('Garage Area tanpa 0')

# Statistik
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())

plt.show()
```

### Analisa
- SalePrice dan GrLivArea setelah di transformasi, distribusinya menjadi normal <br><br>

Untuk GarageArea setelah nilai 0 dihilangkan:
- Mean: 500.76  
- Median: 484.0  
- Standar Deviasi: 185.68
- Distribusi menunjukkan positive skewness ringan (condong ke kanan), namun lebih simetris dibandingkan sebelum nilai nol dihapus.
- Perbedaan antara mean dan median semakin kecil, menandakan skewness berkurang.
- Beberapa nilai ekstrim (outlier) di atas 1000 sqft masih ada, namun tidak terlalu memengaruhi distribusi secara signifikan.
- Menghapus nilai 0 berhasil menghilangkan lonjakan pada bagian paling kiri, menjadikan distribusi lebih representatif untuk rumah yang memiliki garasi.


# 2D Evaluation

## Hubungan antara GrLivArea dan SalePrice

```python
ax = sns.scatterplot(y="SalePrice", x="GrLivArea", data=house_numerical)
plt.show()
```

### ðŸ“Š Scatter Plot: GrLivArea vs SalePrice

- Terdapat hubungan positif yang cukup jelas antara **GrLivArea** (luas bangunan) dan **SalePrice** (harga jual).
- Semakin besar luas bangunan, umumnya harga rumah juga semakin tinggi.
- Data menunjukkan pola linear yang kuat, meskipun terdapat **beberapa outlier**:
  - Rumah dengan **luas sangat besar (>4000 sqft)** namun memiliki harga jual relatif rendah.
  - Juga terdapat rumah dengan **harga sangat tinggi** walau tidak memiliki GrLivArea yang ekstrem.
- Outlier seperti ini bisa berdampak negatif pada model prediktif, dan **perlu ditinjau lebih lanjut** apakah akan:
  - Dihapus
  - Ditandai sebagai data khusus
  - Atau dianalisis lebih dalam konteksnya


```python
# Filter baris dengan GarageArea > 0
filtered_data = house_numerical[house_numerical["GarageArea"] > 0]

# Scatter plot tanpa nilai 0
ax = sns.scatterplot(y="SalePrice", x="GarageArea", data=filtered_data)
plt.title("Scatter Plot SalePrice vs GarageArea (tanpa nilai 0)")
plt.show()
```

### Scatter Plot: SalePrice vs GarageArea (tanpa nilai 0)

Visualisasi ini menunjukkan hubungan antara luas garasi (`GarageArea`) dan harga jual rumah (`SalePrice`), setelah menghapus nilai 0 pada `GarageArea`.

- Terdapat korelasi positif antara luas garasi dan harga jual rumah.
- Data lebih terkonsentrasi pada `GarageArea` antara 200â€“800 sqft.
- Beberapa outlier terlihat pada bagian kanan atas plot, menunjukkan rumah dengan garasi besar dan harga tinggi.
- Menghapus nilai 0 membantu memfokuskan analisis pada rumah-rumah yang memang memiliki garasi.

Visualisasi ini membantu memperjelas pola dan hubungan antara kedua variabel secara lebih akurat.
