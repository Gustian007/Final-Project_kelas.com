# **Data Processing / Feature Engineering**  

## Drop outlier

```python
features = df.drop(columns=['quality']).columns

def remove_outliers_iqr(df, features):
    df_clean = df.copy()
    for col in features:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Hapus data yang outlier di kolom ini
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean

# drop outlier
df_clean = remove_outliers_iqr(df, features)

# reset index
df_clean = df_clean.reset_index(drop=True)
```

# Feature Scaling (Standard Scaler)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Memisahkan fitur dan target
X_clean = df_clean.drop(columns=['quality'])
y_clean = df_clean['quality']

# 
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Menggunakan StandardScaler untuk menormalkan data
scaler = StandardScaler()

# Fit transform dan transfrom
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Menampilkan contoh data yang sudah discaling (beberapa baris pertama)
print("Contoh data pelatihan setelah di-scaling:")
print(X_train_scaled[:5])
```

# Modelling

## Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Membuat model Linear Regression
linear_reg_model = LinearRegression()

# Melatih model dengan data yang telah discaling
linear_reg_model.fit(X_train_scaled, y_train)

# Memprediksi menggunakan data uji
y_pred = linear_reg_model.predict(X_test_scaled)

# Menghitung MSE dan RÂ²
mse_linear_reg = mean_squared_error(y_test, y_pred)
r2_linear_reg = r2_score(y_test, y_pred)

print(f"Mean Squared Error (Linear Regression): {mse_linear_reg}")
print(f"R-squared (Linear Regression): {r2_linear_reg}")
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Membuat model KNN
knn_model = KNeighborsRegressor(n_neighbors=5)

# Melatih model dengan data yang telah discaling
knn_model.fit(X_train_scaled, y_train)

# Memprediksi menggunakan data uji
y_pred_knn = knn_model.predict(X_test_scaled)

# Menghitung MSE dan RÂ²
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (KNN): {mse_knn}")
print(f"R-squared (KNN): {r2_knn}")
```

## SVR

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Membuat model SVR (Support Vector Regression)
svr_model = SVR(kernel='rbf')  # 'rbf' adalah kernel radial basis function

# Melatih model dengan data yang telah discaling
svr_model.fit(X_train_scaled, y_train)

# Memprediksi menggunakan data uji
y_pred_svr = svr_model.predict(X_test_scaled)

# Menghitung MSE dan RÂ²
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (SVR): {mse_svr}")
print(f"R-squared (SVR): {r2_svr}")
```

## Random Forest (after scaling)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Membuat model Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model dengan data yang telah discaling
rf_model.fit(X_train_scaled, y_train)

# Memprediksi menggunakan data uji
y_pred_rf = rf_model.predict(X_test_scaled)

# Menghitung MSE dan RÂ²
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (Random Forest): {mse_rf}")
print(f"R-squared (Random Forest): {r2_rf}")
```

## Random Forest (without scaling)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Memisahkan fitur dan target dari df_clean
X_clean = df_clean.drop(columns=['quality'])
y_clean = df_clean['quality']

# Pembagian data (80% untuk pelatihan, 20% untuk pengujian)
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Membuat model Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model dengan data tanpa scaling
rf_model.fit(X_train, y_train)

# Memprediksi menggunakan data uji
y_pred_rf = rf_model.predict(X_test)

# Menghitung MSE dan RÂ²
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (Random Forest tanpa scaling): {mse_rf}")
print(f"R-squared (Random Forest tanpa scaling): {r2_rf}")
```

# Hyperparameter Tuning

## Linear Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Menentukan parameter grid untuk alpha (regularisasi)
param_grid = {
    'alpha': [0.1, 1, 10, 100, 1000]
}

# Membuat model Ridge Regression
ridge_model = Ridge()

# Menerapkan GridSearchCV untuk mencari parameter terbaik
grid_search_ridge = GridSearchCV(estimator=ridge_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search_ridge.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best alpha for Ridge Regression: {grid_search_ridge.best_params_['alpha']}")
print(f"Best score: {grid_search_ridge.best_score_}")

# Menggunakan model terbaik untuk prediksi
best_ridge_model = grid_search_ridge.best_estimator_
y_pred_ridge = best_ridge_model.predict(X_test)

# Menghitung MSE dan RÂ²
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (Ridge Regression): {mse_ridge}")
print(f"R-squared (Ridge Regression): {r2_ridge}")
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Menentukan parameter grid untuk n_neighbors dan weights
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 10, 15],
    'weights': ['uniform', 'distance']
}

# Membuat model KNN
knn_model = KNeighborsRegressor()

# Menerapkan GridSearchCV untuk mencari parameter terbaik
grid_search_knn = GridSearchCV(estimator=knn_model, param_grid=param_grid_knn, cv=5, n_jobs=-1, verbose=2)
grid_search_knn.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for KNN: {grid_search_knn.best_params_}")
print(f"Best score: {grid_search_knn.best_score_}")

# Menggunakan model terbaik untuk prediksi
best_knn_model = grid_search_knn.best_estimator_
y_pred_knn = best_knn_model.predict(X_test)

# Menghitung MSE dan RÂ²
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (KNN): {mse_knn}")
print(f"R-squared (KNN): {r2_knn}")
```

## SVR

```python
# Mengaktifkan HalvingGridSearchCV karena ini merupakan fitur eksperimen di sklearn
from sklearn.experimental import enable_halving_search_cv
from sklearn.svm import SVR
from sklearn.model_selection import HalvingGridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Menentukan parameter grid untuk C, epsilon, dan kernel
param_grid_svr = {
    'C': [0.1, 1, 10, 100],
    'epsilon': [0.01, 0.1, 0.2, 0.5],
    'kernel': ['linear', 'rbf']
}

# Membuat model SVR
svr_model = SVR()

# Menerapkan HalvingGridSearchCV untuk mencari parameter terbaik
halving_search_svr = HalvingGridSearchCV(svr_model, param_grid_svr, cv=3, n_jobs=-1, verbose=2, factor=2, random_state=42)
halving_search_svr.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for SVR: {halving_search_svr.best_params_}")
print(f"Best score: {halving_search_svr.best_score_}")

# Menggunakan model terbaik untuk prediksi
best_svr_model = halving_search_svr.best_estimator_
y_pred_svr = best_svr_model.predict(X_test)

# Menghitung MSE dan RÂ²
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (SVR): {mse_svr}")
print(f"R-squared (SVR): {r2_svr}")
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Menentukan parameter grid untuk Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Membuat model Random Forest
rf_model = RandomForestRegressor(random_state=42)

# Menerapkan GridSearchCV untuk mencari parameter terbaik
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=3, n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train, y_train)

# Menampilkan hasil terbaik
print(f"Best parameters for Random Forest: {grid_search_rf.best_params_}")
print(f"Best score: {grid_search_rf.best_score_}")

# Menggunakan model terbaik untuk prediksi
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)

# Menghitung MSE dan RÂ²
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Menampilkan hasil evaluasi
print(f"Mean Squared Error (Random Forest): {mse_rf}")
print(f"R-squared (Random Forest): {r2_rf}")
```
